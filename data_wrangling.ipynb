```python
extra = ['Column1', 'Sentence ID_x', 'Citation_x', 'Text before token_x', 'Text after token_x', 'Sentence ID_y', 'Citation_y', 'Text before token_y', 'Text after token_y', 'Unnamed: 16']
```

```python
df = df.drop(columns=extra, axis=0, inplace=False)
df
```

```python
df = df.rename(columns={'function': 'Function', 'Token ID_x': 'Token_ID_Greek', 'Token ID_y': 'Token_ID_Armenian', 'Token form_x': 'Token_Greek', 'Token form_y': 'Token_Armenian'})
df
```

```python
df['Location'] = df['Gospel'].apply(lambda x: x.split()[1])
df['Gospel'] = df['Gospel'].apply(lambda x: x.split()[0])
df
```

```python
def get_where_found(gr, am):
    out_gr = 0
    out_am = 0
    if pd.notnull(gr):
        out_gr = 1
    if pd.notnull(am):
        out_am = 1
    if out_gr == 0 and out_am == 1:
        return 'only_Arm'
    elif out_gr == 1 and out_am == 0:
        return 'only_Gr'
    elif out_gr == 1 and out_am == 1:
        return 'both_langs'

df['where_found'] = df.apply(lambda x: get_where_found(x.Token_ID_Greek, x.Token_ID_Armenian), axis=1)
df
```

```python
def get_where_found(x):
    if pd.notnull(x):
        return 1
    else:
        return 0

df['Greek_found'] = df['Token_ID_Greek'].apply(lambda x: get_where_found(x))
df['Armenian_found'] = df['Token_ID_Armenian'].apply(lambda x: get_where_found(x))
df
```
